```python
#Print i as long as i is less than 6.


i = 1
while i < 6 :
    print(i)
i = i+1 
```

```python
#Stop the loop if i is 3.


i = 1
while i < 6:
    if i == 3:
        break
    i = i + 1
    print(i) 
```

```python
#In the loop, when i is 3, jump directly to the next iteration.


i = 0
while i < 6:
    i = i + 1
    if i == 3:
        continue 
    print(i)
```

```python
#Print a message once the condition is false.


i = 1
while i < 6:
    print(i)
    i = i + 1
    print("i is no longer less than 6")
```
